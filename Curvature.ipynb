```python
from scipy.optimize import root_scalar
import numpy as np
import pandas as pd

# Define a function to calculate curvature
def calculate_curvature(x_coefficients, y_coefficients, z_coefficients):
    xdp = 0
    ydp = 0
    zdp = 0
    xrp = 0
    yrp = 0
    zrp = 0

    for i in range(10):
        xrp = xrp + x_coefficients[i] * (9 - i)
        yrp = yrp + y_coefficients[i] * (9 - i)
        zrp = zrp + z_coefficients[i] * (9 - i)
        if i != 9:
            xdp = xdp + x_coefficients[i] * (9 - i) * (8 - i)
            ydp = ydp + y_coefficients[i] * (9 - i) * (8 - i)
            zdp = zdp + z_coefficients[i] * (9 - i) * (8 - i)

    rprime = [xrp, yrp, zrp]
    rdp = [xdp, ydp, zdp]

    x = np.cross(rprime, rdp)

    r = np.sqrt(x[0] ** 2 + x[1] ** 2 + x[2] ** 2)

    rp = np.sqrt(xrp ** 2 + yrp ** 2 + zrp ** 2)

    return r / (rp ** 3)

# Read data from the file
with open("output_file.txt", "r") as file:
    data = file.read()

# Split the data into sets of coefficients
sets = data.split("]\n\n")

molecules_csv = []

# Iterate through each set of coefficients
for set_data in sets:
    try:
        # Split the set into lines and extract the molecule code
        lines = set_data.split('\n')
        new_lines = []
        molecule_code = lines[0]
        for x in range(1, 19):
            if "[, " in lines[x]:
                lines[x] = lines[x].replace("[, ", "")
            if "]" in lines[x]:
                lines[x] = lines[x].replace("]", "")
            if "[" in lines[x]:
                lines[x] = lines[x].replace("[", "")
            if x>=3 and x%3 == 0:
                new_line = lines[x-2] + ", " + lines[x-1] + ", " + lines[x]
                new_line = [new_line]
                new_lines.append(new_line)

        #for line in lines:
        #    if "[, " in line:
        #        line = line.replace("[, ", "")
        #    if "]" in line:
        #        line = line.replace("]", "")
        #    line = [line]
        #    new_lines.append(line)
        #molecule_code = lines[0].strip().strip('[],"")

        # Extract and process the coefficients
        for num in new_lines[0]:
            elements = num.split(', ')
            x_coefficients = [float(element) for element in elements]
            #print(x_coefficients)
        for num in new_lines[1]:
            elements = num.split(', ')
            y_coefficients = [float(element) for element in elements]
            #print(y_coefficients)
        for num in new_lines[2]:
            elements = num.split(', ')
            z_coefficients = [float(element) for element in elements]
            #print(z_coefficients)
        for num in new_lines[3]:
            elements = num.split(', ')
            x1_coefficients = [float(element) for element in elements]
            #print(x1_coefficients)
        for num in new_lines[4]:
            elements = num.split(', ')
            y1_coefficients = [float(element) for element in elements]
            #print(y1_coefficients)
        for num in new_lines[5]:
            elements = num.split(', ')
            z1_coefficients = [float(element) for element in elements]
            #print(z1_coefficients)

        #x_coefficients = [float(num) for num in new_lines[1]]
        #y_coefficients = [float(num) for num in lines[5].split()]
        #z_coefficients = [float(num) for num in lines[10].split()]
        #x1_coefficients = [float(num) for num in lines[13].split()]
        #y1_coefficients = [float(num) for num in lines[16].split()]
        #z1_coefficients = [float(num) for num in lines[19].split()]

        # Calculate and print the curvature for each set of coefficients
        curvature_x = calculate_curvature(x_coefficients, y_coefficients, z_coefficients)
        curvature_y = calculate_curvature(x1_coefficients, y1_coefficients, z1_coefficients)

        # Print results with the molecule code
        final_molecule = [molecule_code, curvature_x, curvature_y]
        molecules_csv.append(final_molecule)
    except:
        print("INCORRECT")
        print(sets[0])
    
df = pd.DataFrame(molecules_csv)
df.to_csv('curvatures.csv', index=False, header=False)
```
